```python
import hydra
from metrics import *
import lightning as L
import dataloader
import main
import utils
```

```python
overrides=['mode=sample_eval',
           'eval.checkpoint_path=kuleshov-group/mdlm-owt',
           'data=openwebtext-split',
           'model.length=1024',
           'sampling.predictor=merge',  # ddpm, confidence, adaptive, merge
           'sampling.steps=4000',
           'loader.eval_batch_size=10',
           'sampling.num_sample_batches=1',
           'backbone=hf_dit',
           'sampling.confidence_beta=1.5',
           'sampling.gamma=0.1']  

with hydra.initialize(version_base=None, config_path='configs'):
  config = hydra.compose(config_name='config', overrides=overrides)
  sar_config = hydra.compose(config_name='config', overrides=overrides)
    
L.seed_everything(config.seed)

logger = utils.get_logger(__name__)
tokenizer = dataloader.get_tokenizer(config)

generated_samples = main.generate_samples(config, logger, tokenizer)
```

```python
print(f"⬇ Self-BLEU-4: {compute_self_bleu(generated_samples, n=4):.4f}")
print(f"⬆ Distinct-2:  {compute_distinct_n(generated_samples, n=2):.4f}")
print(f"⬆ Distinct-3:  {compute_distinct_n(generated_samples, n=3):.4f}")
```
